## Paso 1: El Escenario del Problema

```python
# Este es el mapa que nuestros algoritmos explorarán.
graph = {
  'A': ['B', 'C'],
  'B': ['D', 'E'],
  'C': ['F'],
  'D': [],
  'E': ['F'],
  'F': ['G'],
  'G': []
}
```

## Paso 2: El Código Funcional Completo

```python
from collections import deque

# --- Implementación de BFS para encontrar un camino ---
def find_path_bfs(graph, start, goal):
    # La cola almacenará listas, donde cada lista es un camino.
    queue = deque([ [start] ])
    visited = set([start])

    while queue:
        # Obtiene el primer camino de la cola.
        path = queue.popleft()
        node = path[-1] # El último nodo del camino es el actual.

        if node == goal:
            return path # ¡Éxito! Hemos encontrado un camino.

        # Explora los vecinos del nodo actual.
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                new_path = list(path)
                new_path.append(neighbor)
                queue.append(new_path)
    
    return None # Si la cola se vacía, no hay camino.

# --- Implementación de DFS para encontrar un camino (versión recursiva) ---
def find_path_dfs(graph, start, goal, path=None):
    if path is None:
        path = [start]

    if start == goal:
        return path # Caso base: hemos llegado al destino.
    
    for neighbor in graph[start]:
        if neighbor not in path: # Evita ciclos simples
            new_path = find_path_dfs(graph, neighbor, goal, path + [neighbor])
            if new_path:
                return new_path # Propaga la solución encontrada.

    return None # No se encontró camino por esta rama.

# --- Ejecución del Ejercicio ---
start_node = 'A'
goal_node = 'G'
```

## Paso 3: Análisis Detallado de la Ejecución de BFS

```python
print("--- Búsqueda BFS ---")
path_bfs = find_path_bfs(graph, start_node, goal_node)
print(f"Camino encontrado: {path_bfs}")
```

## Paso 4: Análisis Detallado de la Ejecución de DFS

```python
print("\n--- Búsqueda DFS ---")
path_dfs = find_path_dfs(graph, start_node, goal_node)
print(f"Camino encontrado: {path_dfs}")
```
