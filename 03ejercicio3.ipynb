## Paso 1: El Reto

```python
# Grafo que representa la red social.
social_network = {
  'Ana': ['Beto', 'Diana'],
  'Beto': ['Carlos'],
  'Carlos': ['Gabriel'],
  'Diana': ['Elena', 'Fernando'],
  'Elena': [],
  'Fernando': ['Gabriel'],
  'Gabriel': []
}
```

## Paso 2: El Código para Encontrar Conexiones

```python
from collections import deque

# --- Implementación de BFS para encontrar el camino más corto ---
def find_path_bfs(graph, start, goal):
    # La cola almacenará los caminos que estamos explorando.
    queue = deque([ [start] ])
    visited = set([start])
    while queue:
        path = queue.popleft()
        node = path[-1]
        if node == goal:
            return path
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                new_path = list(path)
                new_path.append(neighbor)
                queue.append(new_path)
    return None

# --- Implementación de DFS para encontrar cualquier camino ---
def find_path_dfs(graph, start, goal, path=None):
    if path is None: path = [start]
    if start == goal: return path
    for neighbor in graph[start]:
        if neighbor not in path:
            new_path = find_path_dfs(graph, neighbor, goal, path + [neighbor])
            if new_path: return new_path
    return None

# --- Ejecución del Ejercicio ---
start_node = 'Ana'
goal_node = 'Gabriel'
```

## Paso 3: Análisis Detallado de la Ejecución de BFS

```python
print("--- Buscando la conexión más corta con BFS ---")
path_bfs = find_path_bfs(social_network, start_node, goal_node)
print(f"Conexión encontrada por BFS: {path_bfs}")
```

## Paso 4: Análisis Detallado de la Ejecución de DFS

```python
print("\n--- Buscando una conexión cualquiera con DFS ---")
path_dfs = find_path_dfs(social_network, start_node, goal_node)
print(f"Conexión encontrada por DFS: {path_dfs}")
```
