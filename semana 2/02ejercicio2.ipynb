## Paso 1: El Desafío

```python
# Grafo que representa el mapa del campus.
campus_map = {
  'Entrada': ['Biblioteca', 'Cafeteria'],
  'Biblioteca': ['Laboratorio'],
  'Cafeteria': ['Gimnasio'],
  'Laboratorio': ['Gimnasio'],
  'Gimnasio': ['Salida'],
  'Salida': []
}    
```

## Paso 2: El Código de Navegación

```python
from collections import deque

# --- Implementación de BFS (idéntica al ejercicio anterior) ---
def find_path_bfs(graph, start, goal):
    queue = deque([ [start] ])
    visited = set([start])
    while queue:
        path = queue.popleft()
        node = path[-1]
        if node == goal:
            return path
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                new_path = list(path)
                new_path.append(neighbor)
                queue.append(new_path)
    return None

# --- Implementación de DFS (idéntica al ejercicio anterior) ---
def find_path_dfs(graph, start, goal, path=None):
    if path is None: path = [start]
    if start == goal: return path
    for neighbor in graph[start]:
        if neighbor not in path:
            new_path = find_path_dfs(graph, neighbor, goal, path + [neighbor])
            if new_path: return new_path
    return None

# --- Ejecución del Ejercicio ---
start_node = 'Entrada'
goal_node = 'Salida'
```

## Paso 3: Análisis Detallado de la Ejecución de BFS

```python
print("--- Buscando la ruta más corta con BFS ---")
path_bfs = find_path_bfs(campus_map, start_node, goal_node)
print(f"Ruta encontrada por BFS: {path_bfs}")
```

## Paso 4: Análisis Detallado de la Ejecución de DFS

```python
print("\n--- Buscando una ruta cualquiera con DFS ---")
path_dfs = find_path_dfs(campus_map, start_node, goal_node)
print(f"Ruta encontrada por DFS: {path_dfs}")
```
